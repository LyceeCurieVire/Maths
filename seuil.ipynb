```python
def afficheTermes(n):
    q=mpf(0.75)
    for k in range(n):
        u=-2.5*q**k+2.5
        print(f"u({k})={u}")
```

```python
afficheTermes(10)
```

La fonction afficheTermes affiche ici 10 termes, de $u_0$ à $u_9$.

Algorithme de seuil, avec affichage, pour la suite définie par $u_0=0$ et $u_{n+1}=\dfrac{3u_n+2}{u_n+4}$.

```python
def f(x):
    return (3*x+2)/(x+4)
```

```python
def seuilAvecAffichage(A):
    n=0
    u=mpf(0)
    while u<A:
        print(f"u({n})={u}")
        n=n+1
        u=f(u)
    print(f"u({n})={u}")
```

```python
mp.dps=60
seuilAvecAffichage(mpf(1)-1e-60)
```
